```python
from flask import Flask, request

app = Flask(__name__)

nomessage = ["test", "nomessage"]

@app.route('/')
def index():
    message = request.args.get('p')
    if any(word in message for word in nomessage):
        return 'Contains obscene messages, etc.'
    else:
        return message

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
